```python
import pandas as pd

# Load your CSV file
data = pd.read_csv('clustered_training_data_features_2_clusters_10_clusters.csv')

# Calculate the centroids (mean values) for each cluster
centroids = data.groupby('cluster').mean()

# Extract relevant features for centroid calculation
centroids = centroids[['traveled_distance_pixel', 'elapsed_time', 'straightness', 
                       'direction_of_movement', 'mean_a', 'mean_jerk']]

pd.set_option('display.float_format', lambda x: '%.6f' % x)

# Optionally, round the acceleration and jerk values to a fixed number of decimal places
data['mean_a'] = data['mean_a'].round(6)
data['mean_jerk'] = data['mean_jerk'].round(6)

# Now display the data without scientific notation
print(data.head(10))
centroids
```

```python
import pandas as pd

# Load your CSV file
data = pd.read_csv('Features mean_v,sd_v etc clustered files.csv')

# Calculate the centroids (mean values) for each cluster
centroids = data.groupby('cluster').median()

# Extract relevant features for centroid calculation
centroids = centroids[['mea n_v', 'sd_v',  'mean_vx', 'sd_vx', 
    'mean_vy', 'sd_vy',  'mean_a', 'sd_a', 
    'mean_jerk', 'sd_jerk',  'direction_of_movement']]

pd.set_option('display.float_format', lambda x: '%.6f' % x)

# Optionally, round the acceleration and jerk values to a fixed number of decimal places
data['mean_a'] = data['mean_a'].round(6)
data['mean_jerk'] = data['mean_jerk'].round(6)

# Now display the data without scientific notation
print(data.head(10))
centroids
```
